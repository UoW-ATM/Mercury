# Change and sweep parameters

```python
# Change mercury config parameters
ps = read_mercury_config(config_file='config/mercury_config.toml')
ps['computation__num_iter'] = 2 # number of computations per case study

# Choose scenario and case study to be simulated
scenarios = [-1]
case_studies = [-1]

## Choose some values to be simulated

# The first dictionary will set values to a specific value for all iterations. If you want to sweep
# some parameters, see next dictionary
paras_sc_fixed = {
                'modules__modules_to_load':[]
                }

# The second dictionary will iterate through different values for each parameters. Each value of each parameter
# is simulated against all values of the other parameters.
paras_sc_iterated = {'airports__sig_ct':[15., 17.],
                       }
    
# All values that you do not specify in either of the dictionary will be set to their default value, defined in the scenario
# config file.
```

```python
# Initialise simulation
mercury = Mercury()

# Run and get results
with clock_time():
        results, results_seq = mercury.run(scenarios=scenarios,
                                           case_studies=case_studies,
                                            paras_sc_iterated=paras_sc_iterated,
                                            paras_sc_fixed=paras_sc_fixed,
                                            paras_simulation=ps
                                          )
```

```python
results
```

# Use parametriser

```python
# Choose scenario to be simulated
scenarios = [-1]
case_studies = [-1]

# Use a parametriser to be able to sweep ad-hoc parameters

parametriser_name = 'ParametriserStandard' # this is the default parametriser
parametriser = ParametriserSelector().select(parametriser_name)()

# Note that you can pass directly the name of the parametriser, if it is registred, in the Mercury object below.

## Choose some values to be simulated, only defined by the parametriser
paras_sc_iterated = {'compensation_magnitude_short':[50, 75, 100],
                       }

# You can visualise all parameters defined by the parametriser like this
parametriser.parameters
```

```python
# Initialise simulation
mercury = Mercury()

# Run and get results
with clock_time():
    results, results_seq = mercury.run(scenarios=scenarios,
                                       case_studies=case_studies,
                                        paras_sc_iterated=paras_sc_iterated,
                                        parametriser=parametriser)
```

```python
results
```

# Using modules' parameters

```python
scenarios = [-1]
case_studies = [-1]

## Modules can define new parameters, that can be fixed or iterated upon with the same interface

paras_sc_fixed = {
                'modules__modules_to_load':['nostromo_EAMAN'], # use a module
                'nostromo_EAMAN__command_horizon': 450 # this is a parameter that exists only in this module.
                                                        # The name of the module needs to appear before the
                                                        # parameter itself
                }

paras_sc_iterated = {'nostromo_EAMAN__data_horizon':[800, 1000], # this parameter too exists only in this module
                       }
    
# All values that you do not specify in either of the dictionary will be set to their default value, defined in the scenario
# the config file of the module.
```

```python
# Initialise simulation
mercury = Mercury()

# Run and get results
with clock_time():
    results, results_seq = mercury.run(scenarios=scenarios,
                                       case_studies=case_studies,
                                       paras_sc_fixed=paras_sc_fixed,
                                        paras_sc_iterated=paras_sc_iterated)
```

```python
results
```

# Use custom aggregator

```python
# Aggregators allow use to define the dataframe you get as output. By default the results are aggregated based 
# iterated parameters

# Choose scenario to be simulated
scenarios = [-1]
case_studies = [-1]

# Use an aggregator

## Choose some values to be simulated, only defined by the parametriser
paras_sc_iterated = {'airlines__fuel_price':[0.7, 1.5],
                       }

aggregator_name = 'ResultsAggregatorSimpleReduced' # this is the default aggregator, it computes mean and std over all iterations of thesame values of parameters
aggregator = ResultsAggregatorSelector().select(aggregator_name)(list(paras_sc_iterated.keys()))
```

```python
# Initialise simulation
mercury = Mercury()

# Run and get results
with clock_time():
    results, results_seq = mercury.run(scenarios=scenarios,
                                       case_studies=case_studies,
                                        paras_sc_iterated=paras_sc_iterated,
                                        results_aggregator=aggregator)
```

```python
results
```
